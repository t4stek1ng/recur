```python
import glob
import os
os.environ['CUDA_VISIBLE_DEVICES'] = "0"

from vllm import LLM, SamplingParams, RequestOutput
llm = LLM(
    model="/root/project/models/DeepSeek-R1-Distill-Llama-8B",
    task="generate",
    enable_chunked_prefill=False,
    max_model_len=16384,
    max_seq_len_to_capture=16384,
    gpu_memory_utilization=0.975
)

tokenizer = llm.get_tokenizer()

sampling_params = SamplingParams(
    temperature=0.7,
    max_tokens=16384
)
```

```python
outputs = glob.glob("/root/project/dos/exp/8k_DeepSeek-R1-Distill-Llama-8B/*overthink.txt")

texts = []
for path in outputs:
    with open(path, 'r') as f:
        text = f.read().split('[overthinking]\n')[1]
        texts.append(text)

numbers = [len(tokenizer.tokenize(text)) for text in texts]

sum(numbers)
```

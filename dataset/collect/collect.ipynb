**GSM8K数据集**

```python
import random
import json

with open("/root/project/dos/dataset/gsm8k/test-00000-of-00001.jsonl") as f:
    gsm8k = [json.loads(line) for line in f.readlines()]

gsm8k_sample = random.sample(gsm8k, 20)

with open("/root/project/dos/dataset/ours/gsm8k_sample.jsonl", "w") as f:
    f.writelines([json.dumps(data)+"\n" for data in gsm8k_sample])
```

**MMLU数据集**

```python
import csv
import json
import random

classes = ["college_computer_science_test", "college_physics_test", "econometrics_test"]
fieldnames = ['question', 'A', 'B', 'C', 'D', 'answer']

for c in classes:
    with open(f"/root/project/dos/dataset/mmlu/test/{c}.csv", newline='', encoding='utf-8') as f:
        reader = csv.reader(f)
        rows = random.sample([dict(zip(fieldnames, row)) for row in reader], 20)
        with open(f"/root/project/dos/dataset/ours/{c}.jsonl", "w") as j:
            j.writelines([json.dumps(row)+"\n" for row in rows])
```

**GPQA数据集**

```python
import csv
import json
import random

with open(f"/root/project/dos/dataset/gpqa/gpqa_main.csv", newline='', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        rows = random.sample(list(reader), 20)
        with open(f"/root/project/dos/dataset/ours/gpqa.jsonl", "w") as j:
            j.writelines([json.dumps(row)+"\n" for row in rows])
```

```python
with open("/root/project/dos/dataset/ours/gpqa.jsonl", "r") as f:
    dataset = f.readlines()
    lines = []
    for data in dataset:
        d = json.loads(data)
        lines.append({"question": d["Question"], "answer": d["Correct Answer"], "incorrect answers": [d["Incorrect Answer 1"], d["Incorrect Answer 2"], d["Incorrect Answer 3"]]})

    with open(f"/root/project/dos/dataset/ours/gpqa_sample.jsonl", "w") as j:
        j.writelines([json.dumps(line)+"\n" for line in lines])
```
